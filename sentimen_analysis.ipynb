```python
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import re
```

```python
nltk.download('vader_lexicon')
```

```python
sia=SentimentIntensityAnalyzer()
def calculate_sentiment(review):
  review = re.sub(r'[^\w\s]', '', str(review)).lower()
  sentiment = sia.polarity_scores(str(review))
  return sentiment['compound']

def categorize_sentiment(score,rating):
    if score>0.05:
      if rating>=4:
        return 'Positive'
      elif rating == 3:
        return 'Mixed Positive'
      else:
        return 'Mixed Negative'
    elif score <-0.05:
      if rating <= 2:
        return'Negative'
      elif rating == 3:
        return 'Mixed Negative'
      else:
        return 'Mixed Positive'
    else:
      if rating >=4:
        return 'Positive'
      elif rating <=4:
        return 'Negative'
      else:
        return 'Neutral'
def sentiment_bucket(score):
  if score >= 0.5:
    return '0.5 to 1.0'
  elif 0.0 <= score <0.5:
    return '0.0 to 0.49'
  elif -0.5 <= score <0.0:
    return '-0.49 to 0.0'
  else :
    return '-1.0 to -0.5'
```

```python
df_databus_final['SentimentScore']= df_databus_final['ReviewText'].astype(str).apply(calculate_sentiment)
df_databus_final.head()
```

```python
df_databus_final['SentimentCategory']=df_databus_final.apply(
    lambda row : categorize_sentiment(row['SentimentScore'], row['Rating']),
    axis=1
)
df_databus_final.head()
```

```python
df_databus_final['SentimentBucket']=df_databus_final['SentimentScore'].apply(sentiment_bucket)
df_databus_final.head()
```

```python
jumlah_sentiment_nol=(df_databus_final["SentimentScore"]==0.0).sum()
#hitung total jumlah data
total_data=len(df_databus_final)
#hitung persentase
persentase_sentiment_nol=(jumlah_sentiment_nol/total_data)*100
#tampilkan hasil
print(f"jumlah data dengan SentimenSCore 0.0 adalah {jumlah_sentiment_nol} dari {total_data} ({persentase_sentiment_nol:.2f}%)")
```
